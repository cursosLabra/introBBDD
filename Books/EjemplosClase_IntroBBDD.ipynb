## Uniones de tablas

```python
con.execute("""
  create table cursos (id Text Primary Key, nombre Text, edición Text) ;
""");
```

```python
con.execute("""
 insert into cursos values
      ('C1', 'Álgebra', '2024-25'),
      ('C2', 'Física' , '2024-25'),
      ('C3', 'Física' , '2025-26'),
      ('C4', 'Lógica' , '2025-26');
  """);
```

```python
con.execute("""
  CREATE TABLE notas (
    alumnosId Text,
    cursosId Text,
    Nota Real,
    FOREIGN KEY(alumnosId) REFERENCES alumnos(id),
    FOREIGN KEY(cursosId) REFERENCES cursos(id)
);
""");
```

```python
con.execute("""
 insert into notas values
      ('uo234', 'C1', 8),
      ('uo512', 'C1', 7),
      ('uo545', 'C1', 9),
      ('uo234', 'C2', 4),
      ('uo234', 'C3', 7),
      ('uo545', 'C4', 6);
""")
```
